```python
import random

def simulacion_experimento():
    resultados = []  # Para almacenar los resultados A o S
    aguilas_consecutivas = 0
    soles_consecutivos = 0
    lanzamientos = 0

    while aguilas_consecutivas < 3 and soles_consecutivos < 3:
        
        lanzamiento = random.choice(['A', 'S'])
        resultados.append(lanzamiento)
        lanzamientos += 1

        if lanzamiento == 'A':
            aguilas_consecutivas += 1
            soles_consecutivos = 0
        else:  # lanzamiento == 'S'
            soles_consecutivos += 1
            aguilas_consecutivas = 0

    print(" ".join(resultados), end=" ")
    print(f"- {lanzamientos} lanzamientos")
    return lanzamientos

def promedio_lanzamientos():
    total_lanzamientos = 0
    simulaciones = 20

    for i in range(simulaciones):
        print("{:^10} {:^3}:".format('SIMULACIÓN',i+1), end=" ")
        total_lanzamientos += simulacion_experimento()

    promedio = total_lanzamientos / simulaciones
    print()
    print(f"En promedio fueron necesarios {promedio:.2f} lanzamientos para conseguir que aparezcan consecutivamente 3 veces Aguila o Sol")

# Ejecutar la simulación
promedio_lanzamientos()
```

## Dibujando un pentagono

```python
import turtle
window = turtle.Screen()
t = turtle.Turtle()

t.shape("turtle") #Tortuga


t.color("blue")
t.speed('slowest')

t.begin_fill()
for i in range(5):
    t.forward(100)
    t.left(72)
t.end_fill()

t.exitonclick()
```

### Pentágono relleno

```python
import turtle
window = turtle.Screen()
t = turtle.Turtle()

t.shape("turtle") #Tortuga

t.color("red","blue")
t.speed('normal')

for j in range(10):
    t.right(36)
    t.pendown()
    if j%2==0:
        t.begin_fill()
        for i in range(5):
                t.forward(100)
                t.left(72)
        t.penup()
        t.end_fill()
    else:
        for i in range(5):
                t.forward(100)
                t.left(72)


t.hideturtle()

t.exitonclick()
```

## Polígono
